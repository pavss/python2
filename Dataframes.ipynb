```python
import numpy as np
import pandas as pd
import datetime
```

**DataFrames**

```python
players = ['Mary', 'Margie', 'Isabel', 'Pavan', 'Issac', 'Geo']
```

```python
scores = [50,55,45,60,25,25]
```

```python
game_summary = pd.DataFrame(
    {
        "name" : players,
        "scores" : scores
    }
)
```

```python
game_summary[['name', 'scores']]
```

**ADD COLUMNS**

```python
gender = ['female', 'female', 'female', 'male', 'male', 'male']
game_summary['gender'] = gender
game_summary
```

**Deal with missing values**

```python
sparse = pd.DataFrame({'a' : [1,2,3],
                       'b' : [np.nan, 4, np.nan],
                       'c' : [5, np.nan, np.nan]})
sparse
```

```python
sparse.isna()
```

```python
sparse.fillna(0)
```

```python
sparse.dropna(axis=1)
```

**Group by**

```python
game_summary.groupby('gender').mean()
```

```python
above_40 = game_summary['scores'] > 40
above_40
```

```python
game_summary.groupby(above_40).mean()
```

**Merge and Joining**

```python
age = pd.DataFrame(
    {
        'name' : ['Pavan', 'Mary'],
        'age'  : ['23', '24']
    }
)
```

```python
game_summary.merge(age, left_on='name', right_on='name') 
```

```python
game_summary.merge(age, left_on='name', right_on='name', how='outer')
```
