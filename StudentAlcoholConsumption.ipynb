```python
import pandas as pd
import numpy
```

```python
#assign variable df
csv_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv'
df = pd.read_csv(csv_url)
df.head()
```

```python
stud_alcoh = df.loc[: , "school":"guardian"] #slice dataframe from school until guardian
stud_alcoh.head()
```

```python
capitalizer = lambda x: x.capitalize() #create lambda function to capitalize strings
```

```python
stud_alcoh['Mjob'].apply(capitalizer) #capitalize Mjob and Fjob
stud_alcoh['Fjob'].apply(capitalizer)
```

```python
stud_alcoh.tail() #last elements of data set
```

```python
stud_alcoh['Mjob'] = stud_alcoh['Mjob'].apply(capitalizer) #capitalize Mjob and Fjob
stud_alcoh['Fjob'] = stud_alcoh['Fjob'].apply(capitalizer)
stud_alcoh.tail()
```

```python
def majority(x):  #create majority that returns boolean value
    if x > 17:
        return True
    else:
        return False
```

```python
stud_alcoh['legal_drinker'] = stud_alcoh['age'].apply(majority)
stud_alcoh.head()
```

```python
def times10(x):   #multiply every number of dataset by 10
  if type(x) is int:
    return 10 * x
    return x
```

```python
stud_alcoh.applymap(times10).head(10)
```
