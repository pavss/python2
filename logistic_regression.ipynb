Dataset used: Breast Cancer

```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import io
 
data= pd.read_csv(io.BytesIO(uploaded['breast-cancer.csv']))
data.head(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
sns.countplot(data=data,x='diagnosis')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(data.drop('diagnosis',axis = 1),data['diagnosis'],test_size =0.33 ,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
Y_test
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print('The Confusion matrix is: \n',
      confusion_matrix(pred,Y_test))
print('\n')
print('The Classification report is: \n',
      classification_report(pred,Y_test))
```

This model has an accuracy of 64% in predicting whether cancer diagnosis is malignant or benign.
