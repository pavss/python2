```python
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
%matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'
tips = pd.read_csv(url)

tips.head()
```

```python
del tips['Unnamed: 0']

tips.head()
```

```python
ttbill =  sns.distplot(tips.total_bill);
ttbill.set(xlabel = 'Value', ylabel = 'Frequency', title = 'Total Bill')
sns.despine()
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = tips) # scatter plot showing relationship between total bill and tip
```

```python
sns.pairplot(tips) #shows relationship between total bill, tip and size
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True); Shows relationship between total bill and days
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = tips, jitter = True); #scatter plot with day as y-axis and tip as x-axis
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = tips); #Box plot presenting total bill per day 
```

```python
sns.set(style = "ticks") #Two histograms of tip value based for lunch and dinner
g = sns.FacetGrid(tips, col = "time")
g.map(plt.hist, "tip");
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker") #two scatterplots showing relationship betweeen tip and total bill
g.map(plt.scatter, "total_bill", "tip", alpha = .7)
g.add_legend();
```

```python
sns.boxplot(x = "day", y = "tip", hue = "sex", data = tips); #create a boxplot presenting tips per day differentiation sex(male or female)
```
